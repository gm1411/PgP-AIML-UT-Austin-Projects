```python
#Total number of movie releases per year
tot_movies_by_year = moviedata[['movie id', 'release year']].groupby('release year').count()
tot_movies_by_year.columns=["total releases"]
tot_movies_by_year.head(10)
```

```python
#Total number of movie releases per year per genre
tot_movie_per_yr_per_genre=moviedata[['Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery','Romance', 'Sci-Fi', 'Thriller', 'War', 'Western','release year']].groupby('release year').sum()
tot_movie_per_yr_per_genre.tail(15)
```

```python
#Adding column for total number of release per year to above dataset
tot_movie_per_yr_per_genre['total releases'] = pd.DataFrame(tot_movies_by_year["total releases"])
tot_movie_per_yr_per_genre.tail(15)
```

```python
#Percentage number of releases per year by genre
percent_per_yr_per_gen=tot_movie_per_yr_per_genre.div(tot_movie_per_yr_per_genre['total releases'],axis=0)*100
percent_per_yr_per_gen.tail()
```

```python
#5. Merging both the datasets based on movie id in order to get the movie titles
movie_df = moviedata.merge(ratingdata, on="movie id", how = 'inner')
movie_df.head()
```

```python
#1.Count of number of movie ratings in each rating category
movie_df.groupby('movie id')['rating'].value_counts()
```

```python
#1.Count of number of ratings for movies in descending order
high_numbersof_ratings = movie_df.groupby('movie title').size().sort_values(ascending=False)
high_numbersof_ratings
```

```python
#1.Average ratings of all the movies
movie_avg_ratings = movie_df.groupby('movie title').agg({'rating': [np.size, np.mean]})
movie_avg_ratings
```

```python
#2.Slicing movies with ratings of atleast 100 or more
movies_with_100ratings = movie_avg_ratings['rating']['size'] >= 100
movies_with_100ratings
```

```python
#3.Sort values according to average rating such that movie which highest rating is on top.
movie_avg_ratings[movies_with_100ratings].sort_values([('rating', 'mean')], ascending=False)
```

```python
4.# Selecting top 25 movies based on average rating with number of ratings more than 100
movie_avg_ratings[movies_with_100ratings].sort_values([('rating', 'mean')], ascending=False)[:25]
```
